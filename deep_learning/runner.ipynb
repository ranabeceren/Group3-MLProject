```python
import torch
device = "cuda" if torch.cuda.is_available() else "cpu"
device
```

```python
from models.CNN import CNNBaselineModel

model = CNNBaselineModel(
    input_shape=4,
    hidden_units=32,
    output_shape=1
).to(device)
```

```python
model
```

```python
import torch.nn as nn
from metrics.train_metrics import compute_pos_weight
from torchmetrics.classification import BinaryAccuracy

pos_weight = compute_pos_weight(train_loader, device)
loss_fn = nn.BCEWithLogitsLoss(pos_weight=pos_weight) 
accuracy_fn = BinaryAccuracy(threshold=0.5).to(device)
optimizer = torch.optim.Adam(params=model.parameters(),lr=0.001)
```

```python
from metrics.train_metrics import print_train_time 
from trainers.trainerV2 import train_step, test_step

from timeit import default_timer as timer
from tqdm.auto import tqdm

import torch
device = "cuda" if torch.cuda.is_available() else "cpu"
RANDOM_SEED=42

torch.manual_seed(RANDOM_SEED)
torch.cuda.manual_seed(RANDOM_SEED) # Because our code runs on GPU

# Measure time
start = timer()

# Set epochs
epochs = 300

for epoch in tqdm(range(epochs)): # wraps a progression bar around our epochs
    print(f"Epoch: {epoch}\n---------")

    # Train data
    train_step(model=model,
               data_loader=train_loader,
               loss_fn=loss_fn,
               optimizer=optimizer,
               accuracy_fn=accuracy_fn,
               device=device)
    
    # Test data
    test_step(model=model,
              data_loader=test_loader,
              loss_fn=loss_fn,
              accuracy_fn=accuracy_fn,
              device=device)
    
end = timer()
train_time = print_train_time(start=start,
                              end=end,
                              device=device)
```

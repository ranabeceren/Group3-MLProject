```python
import torch
device = "cuda" if torch.cuda.is_available() else "cpu"
device
```

```python
from models.cnn_baselineV3 import BaseLineCNNModelV3

model = BaseLineCNNModelV3(
    input_shape=4,
    hidden_units=30,
    output_shape=1
).to(device)
```

```python
model
```

```python
from helper_functions.helpers import compute_pos_weight
pos_weight = compute_pos_weight(train_loader, device)
```

```python
import torch.nn as nn
loss_fn = nn.BCEWithLogitsLoss(
    pos_weight=pos_weight
) 
# for binary-class classification problems
optimizer = torch.optim.Adam(
    params=model.parameters(),
    lr=0.001)
```

```python
from helper_functions.helpers import pixel_accuracy, print_train_time, train_stepV2, test_step, dice_score

from timeit import default_timer as timer
from tqdm.auto import tqdm

import torch
device = "cuda" if torch.cuda.is_available() else "cpu"
RANDOM_SEED=42

torch.manual_seed(RANDOM_SEED)
torch.cuda.manual_seed(RANDOM_SEED) # Because our code runs on GPU

# Measure time
start = timer()

# Set epochs
epochs = 300

for epoch in tqdm(range(epochs)): # wraps a progression bar around our epochs
    print(f"Epoch: {epoch}\n---------")

    # Train data
    train_stepV2(model=model,
               data_loader=train_loader,
               loss_fn=loss_fn,
               optimizer=optimizer,
               dice=dice_score,
               device=device)
    
    # Test data
    test_step(model=model,
              data_loader=test_loader,
              loss_fn=loss_fn,
              dice=dice_score,
              device=device)
    
end = timer()
train_time = print_train_time(start=start,
                              end=end,
                              device=device)
```
